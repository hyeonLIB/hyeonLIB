```python
def makedir(path):
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
def downloadfiles(dataframe, path):
    dataframe = dataframe.fillna(0)
    
    count = 0
    for index, row in dataframe.iterrows():
        url = row['downloadUrl1']
        dataName = row['downloadUrl1Filename']
        if url == 0:
            print('using downloadurl2')
            url = row['downloadUrl2']
            dataName = row['downloadUrl2Filename']

        if url != 0:
            file_name = os.path.join(path,str(dataName))
            file_list = os.listdir(path)
            if dataName not in file_list:
                count += 1
                print(url, os.path.join(path,str(dataName)))
                urllib.request.urlretrieve(url, file_name)
    print(count)
```

```python
# path
projectPath = df["projectID"][0]
print(projectPath)
makedir(projectPath)
projectName = df['project'].unique()[0]
projectPath = os.path.join(projectPath, projectName)
makedir(projectPath)
```

```python
list_sample = df['sample'].unique()
print(list_sample)
print(len(list_sample))

a = []

for sample in list_sample:
    if '/' in sample:
        sample_folder = sample.replace('/', '_')
        path = os.path.join(projectPath, sample_folder)
        makedir(path)
    else:
        sample_folder = sample
        path = os.path.join(projectPath, sample_folder)
        makedir(path)
    print(sample)
    if sample != 'None':
        dataframe = df[df['sample']==sample]
        a.append({sample:len(dataframe)})
        downloadfiles(dataframe, path)
print('result : ', a, '\ndataframe length : ', len(df))
```

```python
src = img.imread("Lenna.png")
src = src[:,:,0]

# plt.imshow(src, cmap='gray')
mask = np.array([[1/9, 1/9, 1/9, 1/9, 1/9],
                 [1/9, 1/9, 1/9, 1/9, 1/9],
                 [1/9, 1/9, 1/9, 1/9, 1/9],
                 [1/9, 1/9, 1/9, 1/9, 1/9],
                 [1/9, 1/9, 1/9, 1/9, 1/9]])

mask_size = 5

guard = int((mask_size-1)/2)

dst = np.ones_like(src)

# mask = np.ones((mask_size,mask_size)) / mask_size**2

for h in range(guard, src.shape[0] - guard):
    for w in range(guard,src.shape[1]-guard):
        selected_image = src[h-guard:h+guard+1,w-guard:w+guard+1]
        dst[h,w] = np.sum(selected_image * mask)

plt.imshow(dst, cmap='gray')
plt.show()
```
